```python
# biblioteki
import pandas as pd
from pathlib import Path
import kagglehub
import os
```

```python
path = kagglehub.dataset_download('nikdavis/steam-store-games')

print('Path to dataset files:', path)
```

```python
print(os.listdir(path))
```

```python
df = pd.read_csv(os.path.join(path, 'steam.csv'))
```

```python
df.head()
```

```python
print(df.columns)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['developer'] = df['developer'].fillna('Unknown')
df['publisher'] = df['publisher'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
# wybrane kolumny
columns = [
    'name',
    'release_date',
    'genres',
    'price',
    'developer',
    'publisher',
    'positive_ratings',
    'negative_ratings',
    'average_playtime'
]
```

```python
df = df[columns]
```

```python
df['release_date'].head()
```

```python
# zmiana object na datatime
df['release_date'] = pd.to_datetime(df['release_date'],errors='coerce')
df['release_year'] = df['release_date'].dt.year
```

```python
df['release_year'].head()
```

```python
print(min(df['release_year']))
print(max(df['release_year']))
```

```python
df['total_reviews'] = df['positive_ratings'] + df['negative_ratings']
df['user_score'] = (df['positive_ratings'] / df['total_reviews'] * 100).fillna(0)
```

```python
df['user_score'].head()
```

```python
os.makedirs('dataset', exist_ok=True)
```

```python
df.to_csv('dataset/steam_clean.csv', index=False)
```
